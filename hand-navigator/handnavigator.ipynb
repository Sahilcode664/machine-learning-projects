```python
import cv2 # used for the image and video processing from the inputs
import mediapipe as mp # it is used to check the realtime data from the input images 
import time # it is used for time calling and time checking function

cap=cv2.VideoCapture(0)  # it is the cv2 function for opening the camera for capturing the image as input
mpHands=mp.solutions.hands # it is used to find the landmarks of the hand in given images
hands=mpHands.Hands() #it is used to call the hand function for detecting the hand in image 
mpDraw = mp.solutions.drawing_utils  # it is used to draw the line or object from the given point from landmarks 
pTime=0
cTime=0
while(True):
    sucess,img = cap.read()  # read() is used for raeding the captures files from the VideoCaptures()
    imgRGB = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)  # as cv2 genrate the image in BGR format to convert it into RGB we are using the function cvtColor 
    results=hands.process(imgRGB) # results is the converted image of the hands point into RGB color
    if results.multi_hand_landmarks:
        for handLms in results.multi_hand_landmarks:
            for id,lm in enumerate(handLms.landmark):
                h,w,c = img.shape
                cx,cy =int(lm.x*w),int(lm.y*h)
                print(id,cx,cy)
                if id == 4:
                    cv2.circle(img,(cx,cy),15,(225,0,225),cv2.FILLED)
            mpDraw.draw_landmarks(img,handLms,mpHands.HAND_CONNECTIONS)
    cTime=time.time()
    fps=1/(cTime-pTime)
    pTime=cTime
    cv2.putText(img,str(int(fps)),(10,70),cv2.FONT_HERSHEY_PLAIN,3,(225,0,255), 3 )
    cv2.imshow("Image",img)
    if cv2.waitKey(1) & 0xFF == ord('q'): 
        break 
cap.release() 
cv2.destroyAllWindows() 
    
```
