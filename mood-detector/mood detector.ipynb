```python
import numpy as np
import cv2
from sklearn.neighbors import KNeighborsClassifier
import dlib
detector= dlib.get_frontal_face_detector()
predictor=dlib.shape_predictor("shape_predictor_68_face_landmarks.dat")
data=np.load("face_mood_data.npy")
X=data[:,1:].astype(int)
y=data[:,0]
model= KNeighborsClassifier()
model.fit(X,y)
cap=cv2.VideoCapture(0)
while True:
    ret , frame =cap.read()
    gray=cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    faces=detector(frame)
    for face in faces:
        landmarks= predictor(frame,face)
        expression = np.array([[point.x-face.left() , point.y - face.top()] for point in landmarks.parts()[17:]])
        out=model.predict([expression.flatten()])
        #cv2.putText(frame,str(out[0]),(x,y-10),cv2.FONT_HERSHEY_COMPLEX,2,(0,255,0),2)
        print(out)
    if ret:
        cv2.imshow("my",frame)
    key=cv2.waitKey(1)
    if key== ord("q"):
        break
cap.release()
cv2.destroyAllWindows()
```
